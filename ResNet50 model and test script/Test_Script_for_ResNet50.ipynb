```python
from keras.models import model_from_json
from tensorflow.keras.losses import SparseCategoricalCrossentropy
def predict_Test(image_path,json_file_of_model,weights_path):
  

  #Load the Model from Json File
  json_file = open(json_file_of_model, 'r')
  model_json_c = json_file.read()
  json_file.close()
  model_c = model_from_json(model_json_c)
  #Load the weights
  model_c.load_weights(weights_path)
  #Compile the model
  opt = SGD(lr=1e-4, momentum=0.9)
  model_c.compile(loss=SparseCategoricalCrossentropy(), optimizer=opt,metrics=["accuracy"])
  predicted=[]
  for dir in os.listdir(image_path):
    path_img=os.path.join(image_path,dir)

    #load the image you want to classify
    image = cv2.imread(path_img)
    image = cv2.resize(image, (224,224))
    #cv2_imshow(image)
    #predict the image
    preds = model_c.predict(np.expand_dims(image, axis=0))
    y_hat_predicted_max=np.argmax(preds,axis=1)[0]
    predicted.append([dir,y_hat_predicted_max])
  return predicted

def convert_to_datatafram_save_csv(path_to_savein,predictedImages):
  import pandas as pd
  df = pd.DataFrame(predictedImages, columns = ['image_name', 'label'])
  df.to_csv(path_to_savein,index=False)
```
